```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sber_data = pd.read_csv('data/sber_data.csv')
sber_data.head()
```

```python
sber_data['sub_area'].value_counts()
```

```python
sber_data['price_doc'].max()
```

```python
fig = plt.figure(figsize=(12, 8))

boxplot = sns.boxplot(
    data=sber_data,
    y='ecology',
    x='price_doc',
    orient='h',
    width=0.9,
)
```

```python
fig = plt.figure(figsize=(12, 8))

s_plot = sns.scatterplot(
    y='price_doc', 
    x='kremlin_km', 
    data=sber_data,
    hue=sber_data['ecology'],
)
```
